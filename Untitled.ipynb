```python
import pandas as pd

df = pd.DataFrame({
    "StudyHours": [1, 2, 3, 4, 5],
    "Experience": [5, 4, 4, 3, 2],
    "Money":      [5000, 3000, 20000, 4000, 500]
})

def corr_report(df: pd.DataFrame, method: str, target_column: str):
    df_copy = df.copy()
    X = df_copy.drop(columns=target_column)
    y = df_copy[target_column]
    corr = X.corrwith(y, method=method)
    return corr

corr = corr_report(df, "spearman", "Money")
print(corr)
```
